```python
import os

def count_images_in_directory(directory_path):
    image_extensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp']
    image_count = 0

    for dirpath, dirnames, filenames in os.walk(directory_path):
        for filename in filenames:
            if any(filename.lower().endswith(ext) for ext in image_extensions):
                image_count += 1

    return image_count

directory_path = '/content/drive/MyDrive/Colab Notebooks/Leaf Disease/Tomato_Val/Tomato___healthy'
image_count = count_images_in_directory(directory_path)

print(f"Number of image files in the directory: {image_count}")
```

```python
import os
import shutil

def delete_alternate_images(directory_path):
    file_list = os.listdir(directory_path)
    file_list.sort()  # Sort the file list to ensure consistent deletion of alternates

    for i, filename in enumerate(file_list):
        if i % 2 != 0:
            file_path = os.path.join(directory_path, filename)
            if os.path.isfile(file_path):
                os.remove(file_path)

if __name__ == "__main__":
    directory_path = "/content/drive/MyDrive/Colab Notebooks/Leaf Disease/Tomato_Val/Tomato___healthy"
    # # Create a backup of the directory before deleting alternate images
    # backup_directory = "/content/drive/MyDrive/Colab Notebooks/Leaf Disease/Tomato_Train_2"  # Replace with the desired backup path
    # shutil.copytree(directory_path, backup_directory)

    delete_alternate_images(directory_path)

    print("Alternate images have been deleted, and the directory has been saved.")
```
