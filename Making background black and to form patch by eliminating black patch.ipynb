```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Use cv2_imshow for displaying images in Colab
import matplotlib.pyplot as plt

```

```python
# Define the path to your image in Google Drive
image_path = '/content/drive/MyDrive/Colab Notebooks/Leaf Disease/Tomato_Train/Tomato___Bacterial_spot/00416648-be6e-4bd4-bc8d-82f43f8a7240___GCREC_Bact.Sp 3110.JPG'

# Load the image
image = cv2.imread(image_path)
# Get the dimensions of the image
height, width, _ = image.shape

# Define the border width (adjust this according to your needs)
border_width = 10

# Create a black border around the image
bordered_image = cv2.copyMakeBorder(image, border_width, border_width, border_width, border_width, cv2.BORDER_CONSTANT, value=(0, 0, 0))

# Show the image with the border using cv2_imshow
cv2_imshow(bordered_image)

# Define the lower and upper bounds of the range
lower_bound = 1000
upper_bound = 255

# Iterate through the image and set pixels within the range to black
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        pixel = image[i, j]
        if (pixel >= lower_bound).all() and (pixel <= upper_bound).all():
            image[i, j] = [0, 0, 0]





# Display the modified image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# All patches

```python
# Load the image
image = cv2.imread(image_path)
# Get the dimensions of the image
height, width, _ = image.shape

# Define the border width (adjust this according to your needs)
border_width = 10

# Create a black border around the image
bordered_image = cv2.copyMakeBorder(image, border_width, border_width, border_width, border_width, cv2.BORDER_CONSTANT, value=(0, 0, 0))

# Define the lower and upper bounds of the range
lower_bound = 100
upper_bound = 255

# Iterate through the image and set pixels within the range to black
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        pixel = image[i, j]
        if (pixel >= lower_bound).all() and (pixel <= upper_bound).all():
            image[i, j] = [0, 0, 0]

# Display the modified image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

# Define the patch size
patch_size = 16

# Initialize a counter for the patches
patch_count = 0

# Extract and display 16x16 patches from the bordered image
for i in range(0, height, patch_size):
    for j in range(0, width, patch_size):
        patch = bordered_image[i:i+patch_size, j:j+patch_size]
        plt.figure()
        plt.imshow(cv2.cvtColor(patch, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()
        patch_count += 1

# Print the total number of patches formed
print(f"Total number of patches formed: {patch_count}")
```

# Patches without black pixels

```python
# Load the image
image = cv2.imread(image_path)
# Get the dimensions of the image
height, width, _ = image.shape

# Define the border width (adjust this according to your needs)
border_width = 10

# Creating a black border around the image
bordered_image = cv2.copyMakeBorder(image, border_width, border_width, border_width, border_width, cv2.BORDER_CONSTANT, value=(0, 0, 0))

# Defining the lower and upper bounds of the range
lower_bound = 100
upper_bound = 255

# Iterating through the image and set pixels within the range to black
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        pixel = image[i, j]
        if (pixel >= lower_bound).all() and (pixel <= upper_bound).all():
            image[i, j] = [0, 0, 0]

# Displaying the modified image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
print()
print()
print()
print()
print()

# Defining the patch size
patch_size = 16

# Initializing a counter for the valid patches
valid_patch_count = 0

# Extracting and display 16x16 patches from the bordered image that do not contain black pixels
for i in range(0, height, patch_size):
    for j in range(0, width, patch_size):
        patch = bordered_image[i:i+patch_size, j:j+patch_size]


        if not np.any((patch == [0, 0, 0]).all(axis=2)):      # Check if the patch contains any black pixels
            # Get the size (dimensions) of the patch
            patch_height, patch_width, _ = patch.shape
            print(f"Patch {valid_patch_count + 1} - Size: {patch_height}x{patch_width}")

            plt.figure()
            plt.imshow(cv2.cvtColor(patch, cv2.COLOR_BGR2RGB))
            plt.axis('off')
            plt.show()
            valid_patch_count += 1

# Print the total number of valid patches formed
print(f"Total number of valid patches formed: {valid_patch_count}")
```

```python
import os

# Define the directory where you want to save the patches
save_directory = "/content/drive"

# Create the directory if it doesn't exist
os.makedirs(save_directory, exist_ok=True)

# Extract and save 16x16 patches from the bordered image that do not contain black pixels
for i in range(0, height, patch_size):
    for j in range(0, width, patch_size):
        patch = bordered_image[i:i+patch_size, j:j+patch_size]

        if not np.any((patch == [0, 0, 0]).all(axis=2)):
            # Get the size (dimensions) of the patch
            patch_height, patch_width, _ = patch.shape

            # Save the patch as an image
            patch_filename = f"patch_{valid_patch_count + 1}.png"
            patch_filepath = os.path.join(save_directory, patch_filename)
            cv2.imwrite(patch_filepath, patch)

            print(f"Saved Patch {valid_patch_count + 1} - Size: {patch_height}x{patch_width} at {patch_filepath}")
            valid_patch_count += 1

# Print the total number of valid patches formed
print(f"Total number of valid patches formed: {valid_patch_count}")
```
